```python
def n_tokinazer(text, pos_df):
    tokineze_text = text.map(lambda x: ReToken.tokenize(x.lower()))
    tokineze_pos_text = tokineze_text.map(lambda x: pos_tag(x, tagset='universal'))
    
    for one_txt in tokineze_pos_text:
        pos_df= pd.concat([pos_df,Creating_pos_df(one_txt)],ignore_index=True)
        
    print('done')

    return pos_df
```

```python
def Creating_pos_df(one_txt):
    pos_one_txt_df=pd.DataFrame(index=[0])
    for word in one_txt:
        sign_word=word[1]

        if sign_word in pos_one_txt_df.columns:
            pos_one_txt_df[sign_word].loc[0]+=1

        else:# если тип слова встречается в первый раз, то его добавляем как col в df, заполняем столбец нулями кроме последнего    
            sign_word = pd.Series(name=str(word[1]))
            
            pos_one_txt_df=  pd.concat([pos_one_txt_df, sign_word.to_frame()])
            pos_one_txt_df.rename(columns={'sign_word': word[1]})
            pos_one_txt_df[str(word[1])].loc[0]=1


    return  pos_one_txt_df
```

```python
pos_df=pd.DataFrame(index=[0])

pos_final_df = n_tokinazer(All_df.text, pos_df).drop(labels=[0],axis=0).reset_index().drop(columns=['index'])
```

```python
#значения titles повторяется в fake and true df. Но общих названий у них нет
pos_final_df = pos_final_df.fillna(0)
```

```python
pos_tf_idf_final_df= pos_final_df
for index in range(pos_final_df.shape[0]) :
    sum_value_row = pos_final_df.loc[index]
    if sum_value_row == 0:
        pos_final_df.drop(axis=0, index=index, inplace=True)
        continue
    pos_tf_idf_final_df.loc[index] =pos_final_df.loc[index].map(lambda x: x/sum_value_row  )
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

title_pd= All_df.title
tokineze_title = title_pd.map(lambda x:  word_tokenize(x.lower()))
filtered_sentences = [w for w in tokineze_title if not w.lower() in stop_words]

filtered_sentences_tf_idf = tfidf.fit_transform(filtered_sentences)
```

```python
feature_names = tfidf.get_feature_names()
dense = filtered_sentences_tf_idf.to_array()
title_df = pd.DataFrame(dense, columns=feature_names)
```

```python
final_df= pd.concat([title_df,pos_final_df, All_df['True_news?']], ignore_index=True)
```

```python
X=final_df.drop(columns=['True_news?'])
y=final_df['True_news?']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train, y_val = (X, y, random_state=42, test_size=0.33)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

input_shape= X_train.columns.count()

model = keras.Sequential([
    layers.BatchNormalization(input_shape=input_shape), 
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    
    layers.Dense(512, activation=''relu''),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    
     layers.Dense(1, activation='sigmoid'), 
])
```
